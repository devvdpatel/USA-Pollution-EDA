```python
#make a copy of the dataframe
us_pollution_copy = us_pollution.copy()
```

```python
#parse dates using pd.to_datetime function
us_pollution_copy['Date'] = pd.to_datetime(us_pollution_copy['Date'])
```

```python
us_pollution_copy['Year'] = pd.to_datetime(us_pollution_copy['Date']).dt.year
us_pollution_copy['Month'] = pd.to_datetime(us_pollution_copy['Date']).dt.month
us_pollution_copy['Day'] = pd.to_datetime(us_pollution_copy['Date']).dt.day
us_pollution_copy['DayofWeek'] = pd.to_datetime(us_pollution_copy['Date']).dt.dayofweek
```

```python
from ast import increment_lineno
import seaborn as sns
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
#check the ditribution of the pollutants - SO2, NO2, O3, CO
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

axes[0,0].set_title('Distribution of SO2 AQI')
axes[0,0].hist(x = us_pollution_copy['SO2 AQI'], bins =30)
axes[0,0].set_xlabel('SO2 AQI')
axes[0,0].set_ylabel('Count');

axes[0,1].set_title('Distribution of NO2 AQI')
axes[0,1].hist(x = us_pollution_copy['NO2 AQI'], bins =30)
axes[0,1].set_xlabel('NO2 AQI')
axes[0,1].set_ylabel('Count');

axes[1,0].set_title('Distribution of O3 AQI')
axes[1,0].hist(x = us_pollution_copy['O3 AQI'], bins =30)
axes[1,0].set_xlabel('O3 AQI')
axes[1,0].set_ylabel('Count');

axes[1,1].set_title('Distribution of CO AQI')
axes[1,1].hist(x = us_pollution_copy['CO AQI'], bins =30)
axes[1,1].set_xlabel('CO AQI')
axes[1,1].set_ylabel('Count');

plt.tight_layout(pad=3);
```

The distribution of SO2 and CO2 AQIs has extreme outliers with bulk of the data centered around the values 0 and 20.

The distribution of O3 is normal to the naked eye but we can do further analysis on it.

NO2 has a right-skewed distribution which means all values after 55 can be considered as outliers.

```python
#check the cities with the highest concentration of CO
us_city = us_pollution_copy[['NO2 AQI', 'SO2 AQI', 'CO AQI', 'O3 AQI']].groupby([us_pollution_copy['State'],
 us_pollution_copy['City'], us_pollution_copy['County']]).sum()

us_city = us_city.reset_index()
us_city_conc = us_city.sort_values('CO AQI', ascending = False).head(10)
us_city_conc
```

```python
fig = px.histogram(us_city_conc, x = 'City', y = 'CO AQI', title = 'City with Highest Concentration of CO')
fig.show()

```

```python
#analyzing which 5 states have the highest concentration of all 4 pollutants

#NO2
us_pollution_copy_no2 = us_pollution_copy.groupby('State')['NO2 AQI'].sum().sort_values(ascending = False).head(5)

#CO
us_pollution_copy_co = us_pollution_copy.groupby('State')['CO AQI'].sum().sort_values(ascending = False).head(5)

#SO2
us_pollution_copy_so2 = us_pollution_copy.groupby('State')['SO2 AQI'].sum().sort_values(ascending = False).head(5)

#O3
us_pollution_copy_o3 = us_pollution_copy.groupby('State')['O3 AQI'].sum().sort_values(ascending = False).head(5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,8))

axes[0,0].set_title('Top 5 states with highest NO2 AQI')
sns.barplot(x = us_pollution_copy_no2, y = us_pollution_copy_no2.index, ax = axes[0,0]);

axes[0,1].set_title('Top 5 states with highest CO AQI')
sns.barplot(x = us_pollution_copy_co, y = us_pollution_copy_co.index, ax = axes[0,1]);

axes[1,0].set_title('Top 5 states with highest SO2 AQI')
sns.barplot(x = us_pollution_copy_so2, y = us_pollution_copy_so2.index, ax = axes[1,0]);

axes[1,1].set_title('Top 5 states with highest O3 AQI')
sns.barplot(x = us_pollution_copy_o3, y = us_pollution_copy_o3.index, ax = axes[1,1]);

plt.tight_layout(pad = 3);
```

Both states show a relatively low mean AQI, with values less than 50. But that does mean that there are no health impacts of these pollutants. Daily values have to be studied to understand the concentration levels of these populations.

```python
#Check to see if there is any relationship between the pairs of variables using a correlation matrix
us_pollution_copy_pol = us_pollution_copy[['NO2 AQI', 'SO2 AQI', 'CO AQI', 'O3 AQI']]

correlation_matrix = us_pollution_copy_pol.corr()

plt.figure(figsize = (10,6))
plt.title('Correlation matrix of Pollutants', y = -0.25)
sns.heatmap(correlation_matrix, annot = True)
```

As we can see, CO and NO2 have a correlation coefficient greater than 0.62. This means that as the value of CO increases, the value of NO2 increases. This could be due the fact that CO and NO2 are the primary byproducts when gasoline and diesel are burnt.

#Exploring whether yearly trend between pollutants exists?

```python
#plot aggregate of all four pollutants
f , ax = plt.subplots(figsize = [14,6])
us_pollution_copy.groupby(['Year']).agg({'NO2 AQI' : 'mean',
                                         'SO2 AQI' : 'mean',
                                         'CO AQI' : 'mean',
                                         'O3 AQI' : 'mean'}).plot(colormap = 'rainbow', marker = '*', markersize = '10', ax = ax, linewidth = 2.0)

ax.set_title('Mean Pollutant AQI over Time')
ax.set_xlabel('Average AQI')
ax.set_ylabel('Year')

plt.show()
```

Surprisingly, except for Ozone, the trend of all polutants is downward. This signifies that the air quality in the United States has improved over the years.

#Ozone level analysis - California

```python

us_ozone_cali23.plot( x = 'Month',
                      y = 'O3 AQI',
                      kind = 'box',
                      title = "Concentrations of Ozone by month in California in the year 2023")
```

We can observe that in both 2022 and 2023, there is an increase in the AQI for O3 in April, May and June.

A similar upwards trend in the O3 levels during the months of April, May and June can be seen in Pennsylvania as well. The increase in the cincentrations of Ozone could be because of the variations in weather. The months of April, May and June tend to be the hottest months of the year and Ozone is more readily formed in dry, sunny weather.

#Weekly Trends of Ozone in California and Pennsylvania
Ozone is the only pollutant to reach abnormal levels (AQI over 200) in both California and Pennsylvania. This could be due the fact that both states have high population which results in high emissions of pollutants by cars, power plants, refineries, chemical factories and several other sources. Ozone is a by-product of a chemical reaction between oxides of Nitrogen and volatile organic compounds in the presence of sunlight. It is not a naturally occuring chemical.

```python
unhealthy_o3_cali = us_pollution_copy.query("`O3 AQI` >= 201 and State == 'California'")
unhealthy_o3_penn = us_pollution_copy.query("`O3 AQI` >= 201 and State == 'Pennsylvania'")
```

```python
px.bar(unhealthy_o3_cali, x = unhealthy_o3_cali['DayofWeek'],
                y = unhealthy_o3_cali['O3 AQI'],
                hover_data = ['Day', 'Month', 'Year'],
                labels = {"DayOfWeek" : "Day of Week",
                          "O3 AQI" : "O3 AQI Levels",
                          "Month" : "Month of the Year"},
                color = "DayofWeek",
                title = "Day of the Week with unhealthy levels of O3 in California"
       )
```

```python
px.bar(unhealthy_o3_penn, x = unhealthy_o3_penn['DayofWeek'],
                y = unhealthy_o3_penn['O3 AQI'],
                hover_data = ['Day', 'Month', 'Year'],
                labels = {"DayOfWeek" : "Day of Week",
                          "O3 AQI" : "O3 AQI Levels",
                          "Month" : "Month of the Year"},
                color = "DayofWeek",
                title = "Day of the Week with unhealthy levels of O3 in Pennsylvania"
       )
```

In California, unhealthy levels of Ozone are mostly recorded on Sundays, whereas in Pennsylvania Tuesdays have the most abnormal levels of O3 AQI.

Note: The weather conditions and time of the day data was recorded could be a contributory factor to ozone levels.
